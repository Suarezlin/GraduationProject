```python
%matplotlib inline
import matplotlib # 注意这个也要import一次
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize # import figsize
#figsize(5.31, 3.54) # 设置 figsize
figsize(787/300, 590/300) # 设置 figsize
#figsize(1062/300, 797/300) # 设置 figsize
plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率
# 默认的像素：[6.0,4.0]，分辨率为100，图片尺寸为 600&400
# 指定dpi=200，图片尺寸为 1200*800
# 指定dpi=300，图片尺寸为 1800*1200
# 设置figsize可以在不改变分辨率情况下改变比例
```

```python
myfont = matplotlib.font_manager.FontProperties(fname=r'/Users/suarezlin/Library/Fonts/SimSun.ttf') # 这一行

time = []
tps = []
qps = []
lat = []
tps2 = []
qps2 = []
lat2 = []

with open("/Users/suarezlin/Documents/result_original.txt", 'r') as f:
    lines = f.read().split("\n")
    index = 0;
    
    for line in lines:
        if index != 0:
            time.append(int(line.split(",")[0]))
            tps.append(float(line.split(",")[1]))
            qps.append(float(line.split(",")[2]))
            lat.append(float(line.split(",")[3]))
            tps2.append(float(line.split(",")[4]))
            qps2.append(float(line.split(",")[5]))
            lat2.append(float(line.split(",")[6]))
        index += 1
```

```python
plt.axis([-100, 2000, 80, 500])
plt.plot(time, tps, 'r', linewidth=0.3, label="优化前 TPS 值")
plt.plot(time, tps2, 'b', linewidth=0.3, label="优化后 TPS 值")
plt.xlabel(u'时间/s',  fontproperties=myfont, size=6.5) # 这一段
plt.ylabel(u'TPS',  fontproperties=myfont, size=6.5) # 这一段

plt.annotate(r'优化前TPS值', xy=(1200, 195), xytext=(1400, 220),fontproperties=myfont, fontsize=6.5, arrowprops=dict(facecolor='black', shrink=0.01, width=0.1, headwidth=2, headlength = 2))

plt.annotate(r'优化后TPS值', xy=(1200, 380), xytext=(1400, 400),fontproperties=myfont, fontsize=6.5, arrowprops=dict(facecolor='black', shrink=0.01, width=0.1, headwidth=2, headlength = 2))


#plt.legend(prop={'family' : 'SimSun', 'size'   : 7.5})
#plt.title(station, fontdict={'family' : 'SimSun', 'size'   : 10.5})
plt.yticks(fontproperties = 'SimSun', size = 6.5)
plt.xticks(fontproperties = 'SimSun', size = 6.5)
plt.savefig(fname="tps.pdf",format="pdf", bbox_inches='tight')
```

```python
plt.axis([-100, 2000, 80, 10000])

plt.plot(time, qps, 'r', linewidth=0.3, label="优化前 QPS 值")
plt.plot(time, qps2, 'b', linewidth=0.3, label="优化后 QPS 值")
plt.xlabel(u'时间/s',  fontproperties=myfont, size=6.5) # 这一段
plt.ylabel(u'QPS',  fontproperties=myfont, size=6.5) # 这一段

plt.annotate(r'优化前QPS值', xy=(1200, 4000), xytext=(1400, 4400),fontproperties=myfont, fontsize=6.5, arrowprops=dict(facecolor='black', shrink=0.01, width=0.1, headwidth=2, headlength = 2))

plt.annotate(r'优化后QPS值', xy=(1200, 7600), xytext=(1400, 8000),fontproperties=myfont, fontsize=6.5, arrowprops=dict(facecolor='black', shrink=0.01, width=0.1, headwidth=2, headlength = 2))


#plt.legend(prop={'family' : 'SimSun', 'size'   : 5.5})
#plt.title(station, fontdict={'family' : 'SimSun', 'size'   : 5.5})
plt.yticks(fontproperties = 'SimSun', size =6.5)
plt.xticks(fontproperties = 'SimSun', size = 6.5)
plt.savefig(fname="qps.pdf",format="pdf", bbox_inches='tight')
```

```python
plt.axis([-100, 2000, 80, 1000])

plt.plot(time, lat, 'r', linewidth=0.3, label="优化前延迟时间")
plt.plot(time, lat2, 'b', linewidth=0.3, label="优化后延迟时间")
plt.xlabel(u'时间/s',  fontproperties=myfont, size=6.5) # 这一段
plt.ylabel(u'延迟时间/ms',  fontproperties=myfont, size=6.5) # 这一段

plt.annotate('优化前\n延迟时间', xy=(1300, 600), xytext=(1500, 650),fontproperties=myfont, fontsize=6.5, arrowprops=dict(facecolor='black', shrink=0.01, width=0.1, headwidth=2, headlength = 2))

plt.annotate('优化后\n延迟时间', xy=(1300, 150), xytext=(1500, 200),fontproperties=myfont, fontsize=6.5, arrowprops=dict(facecolor='black', shrink=0.01, width=0.1, headwidth=2, headlength = 2))


#plt.legend(prop={'family' : 'SimSun', 'size'   : 5.5})
#plt.title(station, fontdict={'family' : 'SimSun', 'size'   : 10.5})
plt.yticks(fontproperties = 'SimSun', size = 6.5)
plt.xticks(fontproperties = 'SimSun', size = 6.5)
plt.savefig(fname="lat.pdf",format="pdf", bbox_inches='tight')
```
