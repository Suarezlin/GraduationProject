```python
def wavelet_denoising(data):
    # 小波函数取db4
    db4 = pywt.Wavelet('db4')
        # 分解
    coeffs = pywt.wavedec(data, db4)
        # 高频系数置零
    coeffs[len(coeffs)-1] *= 0
    coeffs[len(coeffs)-2] *= 0
        # 重构
    meta = pywt.waverec(coeffs, db4)
    return meta
```

```python
time = []
voltage = []
peak = []
detect = []
current = []

with open("/Users/suarezlin/Documents/T0050.csv", 'r') as f:
    lines = f.read().split("\n")
    for line in lines:
        time.append(float(line.split(',')[0]))
        voltage.append(float(line.split(',')[1]))
        current.append(float(line.split(',')[4]))
        
# timep = []
voltagep = []
currentp = []

# for i in range(int(len(time)/10)):
    
voltagep = wavelet_denoising(voltage)
currentp = wavelet_denoising(current)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()


ax1.axis([0, 100, 0, 300])
ax2.axis([0, 100, 0, 3])

ax1.plot(time, voltage, 'r', linewidth=0.3, label="开关两端电压")
ax2.plot(time, currentp, 'b', linewidth=0.3, label="回路电流")
ax1.set_xlabel(u'时间/ms',  fontproperties=myfont, size=6.5) # 这一段
ax1.set_ylabel(u'开关两端电压/V',  fontproperties=myfont, size=6.5) # 这一段

ax2.set_xlabel(u'时间/ms',  fontproperties=myfont, size=6.5) # 这一段
ax2.set_ylabel(u'回路电流/A',  fontproperties=myfont, size=6.5) # 这一段


ax1.annotate(r'回路电流', xy=(20, 245), xytext=(25, 260), arrowprops=dict(facecolor='black', shrink=0.01, width=0.1, headwidth=2, headlength = 2))

ax1.annotate(r'开关两端电压', xy=(60, 135), xytext=(70, 150), arrowprops=dict(facecolor='black', shrink=0.01, width=0.1, headwidth=2, headlength = 2))


#plt.legend(prop={'family' : 'SimSun', 'size'   : 7.5})
#plt.title(station, fontdict={'family' : 'SimSun', 'size'   : 10.5})
#fig.yticks(fontproperties = 'SimSun', size = 6.5)
#fig.xticks(fontproperties = 'SimSun', size = 6.5)
fig.savefig(fname="T0050.pdf",format="pdf", bbox_inches='tight')
```
